```python
import numpy as np
import math
import matplotlib.pyplot as plt
z = np.arange(0,200,1)
H_co = 17*np.ones(len(z))
H_co2 = 11*np.ones(len(z))
H_co[0:100] = 11
H_co[100:200]=17
```

```python
P_co1 = 0.0048*math.e**-(z[0:100]/H_co[0:100])
P_co2 = P_co1[99]*math.e**-(z[0:100]/H_co[100:200])
P_co = np.concatenate([P_co1,P_co2],axis=0)
P_co2 = 6*math.e**-(z/H_co2)
test1=np.linspace(11,17,100)
```

```python
plt.figure(figsize=(12,10))
plt.plot(np.log10(P_co),z,label=('CO'))
plt.plot(np.log10(P_co2),z,label=('CO2'));
plt.xlabel('Pressure',fontsize=18)
plt.ylabel('Altitude', fontsize=18)
plt.title('CO and CO2 Pressure Falloff vs Altitude', fontsize=20)
plt.legend(prop={'size':20});
```

```python
plt.figure(figsize=(12,10))
plt.plot(P_co/P_co2,z,label=('CO/CO2 Pressure'));
plt.xlabel('Mixing Ratio', fontsize=18)
plt.ylabel('Altitude', fontsize=18)
plt.title('CO and CO2 Mixing Ratio vs Altitude', fontsize=20)
plt.legend(prop={'size':20});
```
